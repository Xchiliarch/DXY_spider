```python
import wordcloud
import jieba
import re
from imageio import imread
import matplotlib.pyplot as plt
```

```python
stopwords = set()
content = [line.strip() for line in open('StopWords.txt','r',encoding= 'utf-8').readlines()]
stopwords.update(content)
len(stopwords)
```

```python
for i in range(2006,2023):
    file = f"{i}_subject.txt"
    with open(file, "r",encoding = 'utf-8') as f:
        allSentence = f.read()
    seg = jieba.cut(allSentence)

    w = wordcloud.WordCloud(
        width = 4000,height = 3000,max_words = 70,
        stopwords = stopwords,
        background_color="white",font_path = "msyhbd.ttc")
    w.generate(' '.join(seg))
    w.to_file(f'{file}.png')
```

```python
im=[]
for i in range(17):
    im.append(plt.imread(f'{2006+i}_subject.png') )
```

```python
rows = 3
columns = 6
fig = plt.figure()
for i in range(17):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(im[i])
    plt.axis('off')
    plt.title(i+2006)
fig.add_subplot(rows, columns, 18)
plt.imshow(plt.imread('All_wordC.png'))
plt.axis('off')
plt.title("ALL")
plt.savefig('compare.png', dpi = 750, facecolor = 'white',bbox_inches = 'tight')
```
